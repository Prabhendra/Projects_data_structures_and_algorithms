```python
'''
Tower of Hanoi Algorithm Explanation:
The problem involves moving n disks from the source rod to the destination rod using an auxiliary rod, following these rules:

Only one disk can be moved at a time.
A larger disk cannot be placed on top of a smaller disk.
All disks start on the source rod and must be moved to the destination rod.

'''
def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    tower_of_hanoi(n-1, source, destination, auxiliary)
    print(f"Move disk {n} from {source} to {destination}")
    tower_of_hanoi(n-1, auxiliary, source, destination)

# Number of disks
n = 3

# Calling the function with rod names A, B, and C
tower_of_hanoi(n, 'A', 'B', 'C')

'''
How Recursion Works in This Algorithm:

1. Base Case: If there's only one disk (n == 1), move it directly from the source to the destination.

2. Recursive Case:

Move n-1 disks from the source to the auxiliary rod using the destination rod.
Move the nth (largest) disk from the source to the destination.
Move the n-1 disks from the auxiliary rod to the destination using the source rod.
'''

'''
Output for n = 3:

Move disk 1 from A to C
Move disk 2 from A to B
Move disk 1 from C to B
Move disk 3 from A to C
Move disk 1 from B to A
Move disk 2 from B to C
Move disk 1 from A to C
'''

'''
Time Complexity of Tower of Hanoi
The time complexity of the Tower of Hanoi problem is O(2 - 1).

Derivation:

Let () be the number of moves required to solve the problem for n disks.

The recursive relation is:

T(n)=2T(n1)+1

where:
T(n1) represents moving 1 disks to the auxiliary rod.
+1 represents moving the largest disk to the destination.
T(n1) again represents moving the n1 disks from the auxiliary to the destination rod.

Expanding the recurrence:

T(n)=2T(n1)+1
=2(2T(n2)+1)+1
=4T(n2)+2+1
=8T(n3)+4+2+1

Observing the pattern, we get:

T(n)=2**n 1

Final Complexity:
T(n)=O(2**n)
This shows that the Tower of Hanoi grows exponentially with the number of disks, making it infeasible for large values of n.

'''
```

Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1 to word2.

Example 1:

Input: word1 = "horse", word2 = "ros"

Output: 3

Explanation:

horse -> rorse (replace 'h' with 'r')

rorse -> rose (remove 'r')

rose -> ros (remove 'e')


Example 2:

Input: word1 = "intention", word2 = "execution"

Output: 5

Explanation:


intention -> inention (remove 't')

inention -> enention (replace 'i' with 'e')

enention -> exention (replace 'n' with 'x')

exention -> exection (replace 'n' with 'c')

exection -> execution (insert 'u')

```python
'''
1. Dynamic Programming: The solution uses a dynamic programming approach. It creates a table dp where dp[i][j] stores the minimum number of operations needed to convert the first i characters of word1 to the first j characters of word2.

2. Initialization:
The first row of the table is initialized with values from 0 to n (length of word1), representing the cost of deleting characters from word1.
The first column is initialized with values from 0 to m (length of word2), representing the cost of inserting characters into word1.
3. Iteration:
The code iterates through the table, calculating the cost for each cell based on the costs of its neighboring cells.
If the current characters in word1 and word2 match, the cost is the same as the previous diagonal cell (no operation needed).
If the characters don't match, the cost is 1 plus the minimum of the costs of:
Deleting a character from word1 (cell above).
Inserting a character into word1 (cell to the left).
Replacing a character in word1 (diagonal cell).

4. Result: The final result is stored in the bottom-right cell of the table (dp[n][m]), representing the minimum cost to convert the entire word1 to word2.

5. Time and Space Complexity

Time Complexity: O(n * m), where 'n' is the length of word1 and 'm' is the length of word2. We need to fill an n x m table.
Space Complexity: O(n * m) to store the dynamic programming table.
'''

def min_distance(word1, word2):
    """
    Calculates the minimum number of operations (insert, delete, replace)
    needed to convert word1 to word2 using dynamic programming.

    Args:
        word1: The first string.
        word2: The second string.

    Returns:
        The minimum number of operations.
    """

    n = len(word1)
    m = len(word2)

    # Create a table to store distances between prefixes of the words
    dp = [[0 for _ in range(m + 1)] for _ in range(n + 1)]

    # Initialize the first row and column of the table
    for i in range(n + 1):
        dp[i][0] = i
    for j in range(m + 1):
        dp[0][j] = j

    # Fill the table using dynamic programming
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1])

    # The result is in the bottom-right cell of the table
    return dp[n][m]

# Example usage
word1 = "horse"
word2 = "ros"
result = min_distance(word1, word2)
print(result)  # Output: 3
```

Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].

```python
'''
the maximum value in the array using Pythons built-in max() function:
'''
arr = [13, 1, -3, 22, 5]
max_value = max(arr)
print("Maximum value:", max_value)

'''
the maximum value manually using a loop:
'''
arr = [13, 1, -3, 22, 5]
max_value = arr[0]  # Assume first element is max

for num in arr:
    if num > max_value:
        max_value = num

print("Maximum value:", max_value)

'''
Maximum Value in an Array using Recursion

Base Case: If the array has only one element, return that element as the maximum.
Recursive Case: Compare the first element with the maximum of the remaining elements and return the larger one.
'''
def find_max(arr, n):
    # Base case: If there's only one element, return it
    if n == 1:
        return arr[0]

    # Recursive call to find the maximum in the smaller subarray
    return max(arr[n - 1], find_max(arr, n - 1))

# Example array
arr = [13, 1, -3, 22, 5]
n = len(arr)
print("Maximum value:", find_max(arr, n))

'''
Time Complexity Analysis
Let T(n) be the time complexity for an array of size n.
The function makes one recursive call for n-1 elements.

The recurrence relation:
T(n)=T(n1)+O(1)

Expanding the recurrence:
T(n)=T(n1)+O(1)=T(n2)+O(1)+O(1)=...=T(1)+O(n)

Since T(1) = O(1), we get:

T(n)=O(n)

Thus, the time complexity is O(n).

'''

```

Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10].

```python
'''
Base Case: If the array has only one element, return that element as the sum.
Recursive Case: The sum of the array is the first element plus the sum of the remaining elements.

'''

def find_sum(arr, n):
    # Base case: If there's only one element, return it
    if n == 1:
        return arr[0]

    # Recursive call to sum the smaller subarray
    return arr[n - 1] + find_sum(arr, n - 1)

# Example array
arr = [92, 23, 15, -20, 10]
n = len(arr)
print("Sum of array:", find_sum(arr, n))

'''
Let T(n) be the time complexity for an array of size n.
The function makes one recursive call for n-1 elements.
The recurrence relation:
T(n)=T(n1)+O(1)

Expanding the recurrence:
T(n)=T(n1)+O(1)=T(n2)+O(1)+O(1)=...=T(1)+O(n)

Since T(1) = O(1), we get:
T(n)=O(n)

Thus, the time complexity is O(n).

'''
```

Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum of every digit in that number raised to the power of total digits in that number is equal to the number.

Example : 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153 hence 153 is an armstrong
number. (Easy)

Input1 : 153

Output1 : Yes

Input 2 : 134

Output2 : No

```python
'''
Armstrong Number using Recursion
An Armstrong number (or Narcissistic number) is a number where the sum of each digit raised to the power of the number of digits equals the original number.

Base Case: If the number is reduced to 0, return 0.
Recursive Case: Extract the last digit, raise it to the power of total digits, and add it to the recursive sum of the remaining digits.

'''
def count_digits(n):
    """Function to count the number of digits in n"""
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

def armstrong_sum(n, power):
    """Recursive function to calculate the sum of digits^power"""
    if n == 0:
        return 0
    return (n % 10) ** power + armstrong_sum(n // 10, power)

def is_armstrong(n):
    """Check if a number is an Armstrong number"""
    num_digits = count_digits(n)
    return n == armstrong_sum(n, num_digits)

# Example inputs
num1 = 153
print("Yes" if is_armstrong(num1) else "No")  # Output: Yes

num2 = 134
print("Yes" if is_armstrong(num2) else "No")  # Output: No


'''
Time Complexity Analysis

Counting Digits:
count_digits(n)  O(log n) (since it reduces n by a factor of 10 in each step)

Armstrong Sum Calculation:
armstrong_sum(n, power)  O(log n) (since it processes each digit)

Overall Complexity:
O(log n) (both functions run in log time)
'''

'''
Example Execution
Input: 153
count_digits(153)  3
armstrong_sum(153, 3)  1 + 5 + 3 = 153
Output: "Yes"

Input: 134
count_digits(134)  3
armstrong_sum(134, 3)  1 + 3 + 4 = 1 + 27 + 64 = 92  134
Output: "No"
'''
```
